```python
import numpy as np
from sklearn import cluster , datasets
from scipy import misc
```

## USING KMEANS

```python
data = misc.face(gray = True)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.gray()
plt.imshow(data)
plt.show()
```

```python
data.dtype
```

```python
data.shape
```

```python
X = data.reshape(786432,1)
```

```python
from sklearn.cluster import KMeans
```

```python



kmeans1 = KMeans(n_clusters = 5 , init = 'k-means++' , random_state  = 101)
```

```python
img1 = kmeans1.fit_predict(X)
```

```python
img1.shape
```

```python
img1 = img1.reshape(768 , 1024)
```

```python
plt.imshow(img1)
plt.show()
```

## Lets try for colored image

```python
data2 = misc.face()

plt.imshow(data2)
plt.show()
```

```python
kmeans2 = KMeans(n_clusters = 10 , init = 'k-means++' , random_state  = 101)
```

```python
data2.shape
```

```python
X1 = data2.reshape(768*1024 , 3)
```

```python
X1.shape
```

```python
img2  = kmeans2.fit_predict(X1)
```

```python
img2
```

```python
centres = kmeans2.cluster_centers_
```

```python
centres
```

```python
centres.shape
```

```python
centres_int = []
for i in centres:
    for j in i:
        centres_int.append(int(j))
```

```python
centres_int = np.array(centres_int)
centres_int = centres_int.reshape(centres.shape)
```

```python
img3 = []
for i in img2:
    img3.append(centres_int[i])
```

```python
img3 = np.array(img3)
```

```python
img3.shape
```

```python
img3 = img3.reshape(768,1024,3)
```

```python
plt.imshow(img3)
plt.show()
```
